```python
import re
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

STOP = set(ENGLISH_STOP_WORDS)

# domain-specific “stopwords” (noise in product titles)
NOISE = {
    "compatible", "with", "for", "and", "or", "other", "all", "devices", "device",
    "smartphone", "smartphones", "tablet", "tablets", "laptop", "laptops", "desktop", "pc",
    "function", "support", "supports", "technology", "series",
    "made", "india",
    "length", "long", "extended",
    "pack", "piece", "pieces", "set",
    "sync", "data",  # optional; remove if you want “data cable” similarity stronger
}

# colors/material-ish words that usually don’t help similarity
COLORS = {
    "black","grey","gray","white","red","green","blue","glossy","dark","iron","martian"
}

# keep connector terms unified
NORMALIZE = {
    "type-c": "typec",
    "usb-c": "typec",
    "typec": "typec",
    "type": "",      # helps remove stray 'type'
    "c": "",         # optional: prevents lonely “c” tokens
}

def clean_product_title(text: str) -> str:
    if pd.isna(text):
        return ""
    text = str(text).lower()

    # 1) remove bracketed/parenthetical chunks (often specs/colors/pack)
    text = re.sub(r"\(.*?\)", " ", text)

    # 2) remove OS/version compatibility blobs (windows 11/10/7..., mac os..., linux kernel..., xp)
    text = re.sub(r"\bwindows\s*\d+(\.\d+)?(\s*/\s*\d+(\.\d+)?)*\b", " ", text)
    text = re.sub(r"\bmac\s*os\s*\d+(\.\d+)?(-\d+(\.\d+)?)?\b", " ", text)
    text = re.sub(r"\blinux\s*kernel\s*\d+(\.\d+)?(-\d+(\.\d+)?)?\b", " ", text)
    text = re.sub(r"\bxp\b", " ", text)

    # 3) remove apple model/version list tokens
    text = re.sub(r"\b(iphone|ipad|ipod|ios)\b", " ", text)
    text = re.sub(r"\b(13|12|11|x|xs|max|xr|se|8|7|6|5|pro|mini|air)\b", " ", text)

    # 4) remove common measurement/spec patterns
    # numbers + units (80 cm, 1.5m, 3 foot, 32 inches, 2 meter, etc.)
    text = re.sub(r"\b\d+(\.\d+)?\s*(cm|mm|m|meter|metre|meters|metres|ft|feet|foot|inch|inches)\b", " ", text)
    # tech specs (60w, 3a, 18gbps, 480mbps, 1080p, 4k, 60hz, 4.4.3 etc.)
    text = re.sub(r"\b\d+(\.\d+)?\s*(w|a|gbps|mbps|hz)\b", " ", text)
    text = re.sub(r"\b\d{3,4}p\b", " ", text)          # 1080p, 720p
    text = re.sub(r"\b\d+k\b", " ", text)              # 4k
    text = re.sub(r"\b\d+(\.\d+){1,3}\b", " ", text)   # version-like 2.6.18-4.4.3 parts (broad)

    # 5) remove huge “bends lifespan” numbers etc (10000+)
    text = re.sub(r"\b\d+\+\b", " ", text)

    # 6) keep only letters/numbers/spaces, normalize separators
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()

    # 7) normalize connector keywords
    tokens = []
    for w in text.split():
        if w in NORMALIZE:
            w = NORMALIZE[w]
        if not w:  # if normalized to ""
            continue

        if w in STOP:
            continue
        if w in NOISE:
            continue
        if w in COLORS:
            continue
        if len(w) <= 2:
            continue

        tokens.append(w)

    return " ".join(tokens)

df["product_name_clean"] = df["product_name"].apply(clean_product_title)
```

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

STOP = set(ENGLISH_STOP_WORDS)

DOMAIN_NOISE = {
    "compatible", "compatibility", "devices", "device", "enabled", "like", "etc",
    "purchase", "customer", "service", "services", "warranty", "months", "month",
    "date", "contact", "hesitate", "question", "problem", "enjoyment", "ensure",
    "quality", "friendly", "reliable", "respond", "within",
    "offer", "offers", "perfect", "ideal", "best", "comfort", "performance",
    "home", "office",
}

def clean_about_product(text: str) -> str:
    if pd.isna(text):
        return ""

    text = str(text).lower()

    # Replace separators & remove decorative brackets like 【】
    text = text.replace("|", " ")
    text = re.sub(r"[【】\[\]]", " ", text)

    # Remove "compatible with ..." style device lists (biggest noise)
    # This removes phrases that start with "compatible with" up to punctuation/pipe/end.
    text = re.sub(r"compatible with[^.!,;]*", " ", text)

    # Remove explicit apple/android model tokens and common device list words
    text = re.sub(r"\b(iphone|ipad|ipod|ios|android|macbook|chromebook|oneplus|samsung|oppo|vivo|realme|mi)\b", " ", text)
    text = re.sub(r"\b(xr|xs|max|plus|pro|mini|air|se|nano|touch)\b", " ", text)

    # Remove numbers + units/specs (keep the surrounding words like 'fast charging')
    text = re.sub(r"\b\d+(\.\d+)?\s*(mbps|mb/s|gbps|hz|w|a|amp|amps|meter|meters|metre|metres|m|cm|mm|ft|feet|inch|inches|years|year|days|day)\b", " ", text)
    text = re.sub(r"\b\d{3,}\+?\b", " ", text)  # big numbers like 10000+ / 20000

    # Remove punctuation, keep letters/numbers/spaces
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()

    # Token filter
    tokens = []
    for w in text.split():
        if w in STOP:
            continue
        if w in DOMAIN_NOISE:
            continue
        if len(w) <= 2:
            continue
        tokens.append(w)

    return " ".join(tokens)
```

```python
def recommend_by_index(idx, top_n=10, same_category=True):
    query_vec = X[idx]
    scores = cosine_similarity(query_vec, X).flatten()

    # exclude itself
    #scores[idx] = -1

    if same_category:
        cat = df.loc[idx, "top_category"]
        mask = (df["top_category"] == cat).to_numpy()
        scores = np.where(mask, scores, -1)

    top_idx = np.argsort(scores)[::-1][:top_n]

    return df.loc[top_idx, ["product_name_clean", "top_category", "rating", "rating_count"]].assign(similarity=scores[top_idx])
```

```python
def recommend_by_partial_name(query, top_n=10, same_category=True):
    matches = df[df["product_name_clean"].str.lower().str.contains(query.lower(), na=False)]
    if matches.empty:
        return "No match found."
    idx = matches.index[0]
    return recommend_by_index(idx, top_n=top_n, same_category=same_category)
```

```python
def recommend_products_for_user(model, dataset, user_id, product_df, N=5):
    # Extract LightFM internal mappings
    user_id_map, _, item_id_map, _ = dataset.mapping()

    # Check if user_id exists in mapping
    if user_id not in user_id_map:
        raise ValueError(f"User ID {user_id} not found in the dataset mapping.")

    internal_user_id = user_id_map[user_id]
    num_items = len(item_id_map)

    # Predict scores for all items for the user
    scores = model.predict(user_ids=internal_user_id, item_ids=np.arange(num_items))

    # Get top N item indices (internal IDs)
    top_indices = np.argsort(-scores)[:N]  # negative for descending sort

    # Reverse mapping: internal → external movie_id
    inv_item_id_map = {v: k for k, v in item_id_map.items()}
    recommended_product_ids = [inv_item_id_map[i] for i in top_indices]

    # Get movie titles
    recommended_products = product_df[product_df['product_id'].isin(recommended_product_ids)]
    recommended_products = recommended_products.set_index('product_id').loc[recommended_product_ids].reset_index()

    return recommended_products
```

Item based reccomondations

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Item embedding matrix (num_items x embedding_dim)
item_embeddings = model.item_embeddings
#model.item_embeddings is a matrix where each row corresponds to the embedding vector of an item (movie) learned by LightFM during training.

cos_sim_matrix = cosine_similarity(item_embeddings)
item_id_map = dataset.mapping()[2]  # product_id to LightFM index
reverse_item_map = {v: k for k, v in item_id_map.items()}  # index to product_id
def get_top_k_similar_items(product_id, k=5):
    if product_id not in item_id_map:
        print(f"Product ID {product_id} not found in item_id_map.")
        return []

    idx = item_id_map[product_id]
    sim_scores = cos_sim_matrix[idx]

    # Exclude the item itself and get top-k indices
    similar_idx = np.argsort(-sim_scores)  # descending
    similar_idx = similar_idx[similar_idx != idx][:k]  # remove self

    similar_product_ids = [reverse_item_map[i] for i in similar_idx]
    similar_scores = [sim_scores[i] for i in similar_idx]

    return list(zip(similar_product_ids, similar_scores))
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# user-product rating matrix
R = df_collaborative.pivot_table(
    index="user_id",
    columns="product_id",
    values="rating",
    aggfunc="mean",
    fill_value=0
)

# user-user similarity
U_sim = cosine_similarity(R)
U_sim = pd.DataFrame(U_sim, index=R.index, columns=R.index)

def cf_scores_for_user(user_id, k=20):
    """Return CF predicted scores for all products for a given user."""
    if user_id not in R.index:
        raise ValueError("user_id not found")

    sims = U_sim[user_id].drop(user_id).sort_values(ascending=False).head(k)
    neighbor_ids = sims.index
    weights = sims.values

    # weighted sum of neighbors' ratings
    neighbor_ratings = R.loc[neighbor_ids]                 # k x num_products
    pred = np.dot(weights, neighbor_ratings.values) / (weights.sum() + 1e-9)

    return pd.Series(pred, index=R.columns, name="CF_score")
```

```python
df_collaborative["content"] = df_collaborative["product_name_clean"].fillna("") + " " + df_collaborative["about_product_final"].fillna("")

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# One row per product for content model
prod_text = df_collaborative.drop_duplicates("product_id")[["product_id", "content"]].set_index("product_id")

tfidf = TfidfVectorizer(min_df=2, ngram_range=(1,2))
X = tfidf.fit_transform(prod_text["content"])  # products x features

# Map product_id -> row index in X
prod_index = pd.Index(prod_text.index)
prod_to_row = {pid:i for i, pid in enumerate(prod_index)}

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def content_scores_for_user(user_id, top_m=10):
    user_hist = df_collaborative[df_collaborative["user_id"] == user_id][["product_id","rating"]]
    if user_hist.empty:
        raise ValueError("No history for this user")

    user_hist = user_hist.sort_values("rating", ascending=False).head(top_m)

    rows = [prod_to_row[p] for p in user_hist["product_id"] if p in prod_to_row]
    if not rows:
        raise ValueError("User history products not found in content index")

    # ✅ convert mean vector to normal ndarray
    user_vec = np.asarray(X[rows].mean(axis=0))

    sims = cosine_similarity(user_vec, X).flatten()
    return pd.Series(sims, index=prod_index, name="Content_score")
```

```python
def minmax(s):
    s = s.astype(float)
    return (s - s.min()) / (s.max() - s.min() + 1e-9)

def hybrid_recommend(user_id, n=10, alpha=0.6, k=20, top_m=10, min_ratings=16000):
    cf = cf_scores_for_user(user_id, k=k)
    content = content_scores_for_user(user_id, top_m=top_m)

    # align indexes (products)
    scores = pd.concat([cf, content], axis=1).fillna(0)

    # normalize both to 0..1
    scores["CF_norm"] = minmax(scores["CF_score"])
    scores["Content_norm"] = minmax(scores["Content_score"])

    # score fusion
    scores["Fusion"] = alpha * scores["CF_norm"] + (1 - alpha) * scores["Content_norm"]

    # remove already-rated products
    seen = set(df_collaborative.loc[df_collaborative["user_id"] == user_id, "product_id"])
    scores = scores.drop(index=[p for p in seen if p in scores.index], errors="ignore")

    # popularity filter (optional — your existing idea)
    pop_ok = popular_products.set_index("product_id")  # from your code
    scores = scores.join(pop_ok[["avg_rating", "rating_count"]], how="left")
    scores = scores[scores["rating_count"].fillna(0) >= min_ratings]

    return scores.sort_values("Fusion", ascending=False).head(n).reset_index().rename(columns={"index":"product_id"})
```
