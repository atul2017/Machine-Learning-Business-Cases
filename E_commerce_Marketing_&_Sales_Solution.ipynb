Invoice value calculation -
Invoice Value = ((Quantity * Avg_price) * (1 - Discount_pct) * (1 + GST)) + Delivery_Charges

Tables invloved in this calculation -
  Quantity - Online_Sales data
  Avg Price - Online sales data
  Discount percentage - Discount_Coupon
  GST = Tax_amount

```python

# Convert to ordered categorical so it sorts chronologically
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df_online_sales_data['Month'] = pd.Categorical(df_online_sales_data['Month'], categories=month_order, ordered=True)

customers_by_month = (
    df_online_sales_data.groupby('Month')['CustomerID']
      .apply(set)
      .reset_index(name='Customers')
).sort_values('Month')
print(customers_by_month)
```

```python
retention_results = []

for i in range(len(customers_by_month) - 1):
    month_x = customers_by_month.loc[i, 'Customers']
    month_x1 = customers_by_month.loc[i + 1, 'Customers']

    returning_customers = month_x.intersection(month_x1)
    retention_rate = len(returning_customers) / len(month_x)

    retention_results.append({
        'Month': f"{customers_by_month.loc[i, 'Month']} → {customers_by_month.loc[i + 1, 'Month']}",
        'Customers_in_Month_X': len(month_x),
        'Returning_in_X+1': len(returning_customers),
        'Monthly_Retention_Rate': round(retention_rate, 2)
    })

retention_df = pd.DataFrame(retention_results)
print(retention_df)
```

Let's see the revenue generated by cutomers retention versus new customers

```python
# Ensure months are ordered
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']
df_Invoice['Month'] = pd.Categorical(df_Invoice['Month'], categories=month_order, ordered=True)

customers_by_month = (
    df_Invoice.groupby('Month')['CustomerID']
    .apply(set)
    .reset_index(name='Customers')
)

```

```python
results = []

for i in range(len(customers_by_month) - 1):
    month_x = customers_by_month.loc[i, 'Customers']
    month_x1 = customers_by_month.loc[i + 1, 'Customers']

    month_name = customers_by_month.loc[i + 1, 'Month']  # next month’s name

    # Retained = in both months
    retained = month_x.intersection(month_x1)

    # New = only in next month
    new_customers = month_x1 - month_x

    # Filter data for next month
    df_next = df_Invoice[df_Invoice['Month'] == month_name]

    # Calculate revenue
    retained_revenue = df_next[df_next['CustomerID'].isin(retained)]['Invoice_Value'].sum()
    new_revenue = df_next[df_next['CustomerID'].isin(new_customers)]['Invoice_Value'].sum()
    total_revenue = df_next['Invoice_Value'].sum()

    results.append({
        'Month': month_name,
        'Retained_Customers': len(retained),
        'New_Customers': len(new_customers),
        'Retained_Revenue': retained_revenue,
        'New_Revenue': new_revenue,
        'Total_Revenue': total_revenue,
        'Retention_Revenue_%': round((retained_revenue / total_revenue) * 100, 2) if total_revenue > 0 else 0,
        'New_Revenue_%': round((new_revenue / total_revenue) * 100, 2) if total_revenue > 0 else 0
    })

retention_revenue_df = pd.DataFrame(results)

retention_revenue_df
```

```python
total_row = {
    'Month': 'Total',
    'Retained_Revenue': retention_revenue_df['Retained_Revenue'].sum(),
    'New_Revenue': retention_revenue_df['New_Revenue'].sum(),
    'Total_Revenue': retention_revenue_df['Total_Revenue'].sum()
}

# Append to the DataFrame
retention_revenue_df = pd.concat(
    [retention_revenue_df, pd.DataFrame([total_row])],
    ignore_index=True
)

retention_revenue_df = retention_revenue_df.round({
    'Retained_Revenue': 2,
    'New_Revenue': 2,
    'Total_Revenue': 2
})
retention_revenue_df = retention_revenue_df.drop_duplicates()


retention_revenue_df.sort_values('Month')
```

Customers segmentation in platinum, gold and silver based on RFM (Recency, Frequency & Monetry) techniques

```python
df_online_sales_data['Transaction_Date'] = pd.to_datetime(df_online_sales_data['Transaction_Date'])


df_online_sales_data['Revenue'] = df_online_sales_data['Quantity'] * df_online_sales_data['Avg_Price']

snapshot_date = df_online_sales_data['Transaction_Date'].max() + pd.Timedelta(days=1)


rfm = df_online_sales_data.groupby('CustomerID').agg({
    'Transaction_Date': lambda x: (snapshot_date - x.max()).days,  # Recency
    'Transaction_ID': 'nunique',                                  # Frequency
    'Revenue': 'sum'                                              # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 3, labels=[3, 2, 1]).astype(int)
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 3, labels=[1, 2, 3]).astype(int)
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 3, labels=[1, 2, 3]).astype(int)


rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

rfm.head(10)
```

```python
def segment_customer(score):
    if score >= 8:
        return 'Platinum'
    elif score >= 5:
        return 'Gold'
    else:
        return 'Silver'

rfm['Segment'] = rfm['RFM_Score'].apply(segment_customer)
rfm.head(10)
```

Do customers who use coupons have a different average transaction value compared to those who do not? Conduct a statistical test to validate this hypothesis. What implications does this have for the company's discount and coupon strategies?




Hypothesis -

X1 = mean transaction value for customers who used the coupon

X2 = mean transaction value for customers who did no use the coupon


H0 (Null Hypothesis) There is no significant difference in the average transaction value between coupon users and non coupon user
      X1=X2


H1 (Alternative Hypothesis) There is a significant difference in the average transaction value
      X1 != X2



```python
from scipy.stats import ttest_ind

from scipy.stats import mannwhitneyu

coupon_users = df_Invoice[df_Invoice['Coupon_Status'] == 'Used']['Invoice_Value']
non_coupon_users = df_Invoice[ (df_Invoice['Coupon_Status'] == 'Not Used') | (df_Invoice['Coupon_Status'] == 'Clicked')]['Invoice_Value']

t_stat, p_value = ttest_ind(coupon_users, non_coupon_users, equal_var=False, nan_policy='omit')

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.5f}")
```

Since the p>.05 we cannot ignore the null hypothesis. This concludes there is no significant difference in the average transaction value between coupon users or non coupon users.

 Does customer tenure impact purchase frequency? Analyze the relationship between customer tenure and purchase frequency. How can this insight be used to improve customer engagement and retention strategies?


Let's analyze the purchase freqeuncy with customers tenure to get some insights about the customers behaviour based on their tenure.


Null hypothesis - Customers tenure does not impact their purchase frequency.

Alternative Hypothesis - Customers tenure does significantly impact their purchase frequency.


```python
import pandas as pd

# Find first and last purchase per customer
customer_tenure = df_online_sales_data.groupby('CustomerID')['Transaction_Date'].agg(['min', 'max', 'count']).reset_index()
# Calculate tenure (in months) and frequency
customer_tenure['Tenure_Months'] = ((customer_tenure['max'] - customer_tenure['min']).dt.days / 30).round(1)
customer_tenure['Purchase_Frequency'] = customer_tenure['count'] / (customer_tenure['Tenure_Months'] + 1)
```

```python
customer_tenure = customer_tenure[customer_tenure['Tenure_Months'] > 0]

customer_tenure['Tenure_Group'] = pd.qcut(customer_tenure['Tenure_Months'], 3, labels=['Low', 'Medium', 'High'], duplicates='drop')
```

```python
#Using the one way ANOVA test

from scipy.stats import f_oneway

groups = [customer_tenure[customer_tenure['Tenure_Group'] == g]['Purchase_Frequency'] for g in customer_tenure['Tenure_Group'].unique()]
f_stat, p_val = f_oneway(*groups)

print(f"F-statistic = {f_stat:.3f}, P-value = {p_val:.5f}")
```

The p-value<.05 that means customer tenure strongly affects purchase frequency.The strategic initiative could be creating loyality program where customers have benefits like reward program based on their time and **purchases**.
